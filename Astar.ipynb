```python
def h_manhatan(x,y,x_goal,y_goal):
    """return ((x-x_goal)**2+(y-y_goal)**2)**1/4"""  
    return (abs(x-x_goal)+abs(y-y_goal)) 


from heapq import heappop, heappush
import string
from IPython.display import clear_output
from time import sleep
```

```python
from heapq import heappop, heappush
"""
look for adjuency list represntation for graphs https://en.wikipedia.org/wiki/Adjacency_list
"""
def read_maze(file_path):
    f = open(file_path, "r")
    text=f.read()
    """
    split lines in array of lines
    """
    lines=text.split('\n')
    """
    store points as keys in dict and their childs as set and add each child in set  using adjuency list convention
    """
    points={(x,y): set() for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
    """
    store points visited in order not to get stuck in loops with defualt for not visited as false
    """
    points_visited={(x,y): False for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
    """
    store distances  in order to track cost of reaching to points with infity as defulat
    """
    distances={(x,y): float('inf') for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
   
    """
    when finds "p" store start location in start
    """
    start=(0,0)
    """
    store goals in goals array as when scanner finds "." adds its locations in array  
    """
    print(":height")
    
    print(len(lines))
    print("width")
    print(len(lines[0]))
    goals=[]
    for y in range(0,len(lines)) :
        for x in range(0,len(lines[0])):
            if len(lines[y])>1:
                if lines[y][x]!="%":
                    if y>0:
                    
                        if lines[y-1][x]!="%":
                            points[(x,y)].add((x,y-1))
                    if x>0:
                        if lines[y][x-1]!="%":
                            points[(x,y)].add((x-1,y))
                    if y<len(lines)-2:
                        if lines[y+1][x]!="%":
                            points[(x,y)].add((x,y+1))
                    if x<len(lines[0])-2:
                        if lines[y][x+1]!="%":
                            points[(x,y)].add((x+1,y))
                if lines [y][x]=='.':
                    goals.append((x,y))
                if lines [y][x]=='P':
                    start=(x,y)
                
    to_be_delted=[]
    """
    delte not used points
    """
    
    sorted_goals=[]
    for goal in goals :
        heappush(sorted_goals,(h_manhatan(start[0],start[1],goal[0],goal[1]),goal))
    goals=[]
    i=1
    while sorted_goals:
        

        
        r,g=heappop(sorted_goals)

        goals.append(g)
    
    paths={(x,y): [(start)] for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
    f={(x,y): 0 for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
    h={(x,y): h_manhatan(x,y,goals[0][0],goals[0][1]) for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
    h_list={}
    for i  in range(0,len(goals)):
        h={(x,y): h_manhatan(x,y,goals[i][0],goals[i][1]) for x in range(0,len(lines[0]))  for y in range(0,len(lines)) }
        h_list[(goals[i][0],goals[i][1])]=h
    for key in points.keys():
        if len(points[key])==0:
            to_be_delted.append( key)
       
    for key in to_be_delted:
        del points[key]
        del distances[key] 
        del points_visited[key]
        del paths[key]
        del f[key]
        del h[key]
        
    return {"start":start,"goals":goals , "points":points ,"distances":distances,"lines":lines,"visited":points_visited,"paths":paths,"f":f,"h_list":h_list
        } 
"""
testing__________________________

"""
"""maze=read_maze("bigDots.txt")
print("start====================================================================")
print(maze["start"])
print("goals====================================================================")

print(maze["goals"])
print("points===================================================================")

print(maze["points"])
print("distances===============================================================")

print(maze["distances"]) 
print("Maze drawing===============================================================")
for line in maze["lines"]:
    
    print(line)            
maze=read_maze("bigMaze.txt")
 """
```
